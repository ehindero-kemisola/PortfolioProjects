### EDA

##### Analyzing different Shipping Modes, Segments and Categories in the dataset

```python
supermarket["Ship Mode"].value_counts()
```

```python
sns.countplot(x=supermarket["Ship Mode"])
```

```python
supermarket["Segment"].value_counts()
```

```python
sns.countplot(x=supermarket["Segment"])
```

```python
supermarket["Category"].value_counts()
```

```python
sns.countplot(x=supermarket["Category"])
```

#### From the above plot we can conclude that Office Supplies Category has highest number of sales.

```python
supermarket["Sub-Category"].value_counts()
```

```python
plt.figure(figsize=(15,15))
plt.pie(supermarket["Sub-Category"].value_counts(), labels= supermarket["Sub-Category"].value_counts().index, autopct ="%2f")
plt.show()
```

#### Here, Sub-Category with highest sale is Binder, follwed by Paper and Furnishings as second and third respectively

```python
state_profit=supermarket.groupby(["State"])["Profit"].sum().nlargest(20)
```

```python
state_profit
```

```python
plt.figure(figsize=(15,8))
state_profit.plot.bar()
```

```python
sns.lineplot(data=supermarket, x="Discount", y= "Profit")
```

#### When discount increases, profit decreases

```python
supermarket.plot(kind="scatter",x="Sales",y="Profit", c="Discount", colormap="Set1",figsize=(10,10))
```

#### In this scatter, plot we see that more sales does not mean more profit. It depends on discount as well. When Sales is high and there is low discount, the profit margin is higher.

```python
supermarket1= supermarket.groupby("State")[["Sales","Profit"]].sum().sort_values(by="Sales", ascending=False)
supermarket1[:].plot.bar(color = ["Green","Red"], figsize=(20,12))
plt.title("Profit-Loss and Sales across States")
plt.show()
```

#### California and New York generate more profit compared to the other states

### Profit-Loss and Sales across Region

```python
supermarket1= supermarket.groupby("Region")[["Sales","Profit"]].sum().sort_values(by="Sales", ascending=False)
supermarket1[:].plot.bar(color = ["Blue","Red"], figsize=(10,7))
plt.title("Profit-Loss and Sales across Region")
plt.show()
```

### Conclusion
1. The western region generates highest profit.
2. California, NewYork and Washington generates the most sales compared to the other places.
3. The central region generates lowest profit.
4. Texas, Pennsylvenia, Florida, Illinois, Ohio and some other states are generating loss with high sale. So we need to give some attention towards them.
    Therefore, we have to work more on California and New York. Increase the sales in these states by reducing sales in states like Texas, Florida, Ohio. By decreasing the discount rates in Central region we can increase the profit. Finally we should increase the sale of Office Supplies category as they contribute more.
