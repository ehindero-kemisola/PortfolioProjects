    Monthly average house prices
    Yearly number of houses sold

#### The data used is from year 1995 to 2019 of each different area

```python
# Convert the Datatype of Date column to Date-Time format

housing.dtypes
```

```python
housing.date = pd.to_datetime(housing.date)
```

```python
housing.dtypes
```

```python
# Add a new column 'year' in the dataframe, which contains year only

housing['year'] = housing.date.dt.year
```

```python
housing
```

```python
# Add a new column 'month' as 2nd column in the dataframe, which contains month only

housing.insert(2, 'month', housing.date.dt.month)
```

```python
housing
```

```python
# Remove the columns 'year' and 'month' from the dataframe

housing.drop(['month', 'year'], axis=1, inplace = True)
```

```python
housing.head()
```

```python
# Show all the records where 'No. of Crimes' is 0. And how many such records are there

housing[housing.no_of_crimes == 0]
```

```python
len(housing[housing.no_of_crimes == 0])
```

```python
# What is the maximum & minimum 'average_price' per year in england

housing['year'] = housing.date.dt.year
```

```python
df1 = housing[housing.area == 'england']
df1
```

```python
df1.groupby('year').average_price.max()
```

```python
df1.groupby('year').average_price.min()
```

```python
df1.groupby('year').average_price.mean()
```

```python
# What is the Maximum & Minimum No. of Crimes recorded per area

housing.groupby('area').no_of_crimes.max()
```

```python
housing.groupby('area').no_of_crimes.min().sort_values(ascending= True)
```

```python
# Show the total count of records of each area. where average price is less than 100000

housing[housing.average_price < 100000].area.value_counts()
```
